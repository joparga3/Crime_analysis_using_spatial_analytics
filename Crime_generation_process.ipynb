# Generation of aggregates

## Total counts

* Combine 2015-2016 and 2017-2018 records
    * Take count of the total and each individual crimes
    * Counts might be interesting, but on their own, can be actually misleading for a model. When counts on their own are used, it normally acts as a proxy for population density (ie, higher areas of crimes occur in densely populated areas).
    * As mentioned, we will generate these averages, but in the following section normalise them using the census populations.

```r
crimes_subset_clean$date_year = as.numeric(substr(crimes_subset_clean$date_created,1,4))
crimes_subset_clean = crimes_subset_clean %>% filter(date_year >= 2015)
crimes_subset_clean$date_created2 = ifelse(crimes_subset_clean$date_year %in% c(2015,2016),'2015_2016'
                                          , ifelse(crimes_subset_clean$date_year %in% c(2017,2018),'2017_2018','Other'))
```

```r
aggregates1_1 = crimes_subset_clean %>% group_by(lsoa_code, date_created2, crime_type) %>% summarise(cnt=n())
aggregates1_1 = aggregates1_1 %>% spread(key=crime_type, value = cnt)

aggregates1_2 = crimes_subset_clean %>% group_by(lsoa_code, date_created2) %>% summarise(Total=n())

aggregates1_3 = inner_join(aggregates1_2, aggregates1_1, by = c('lsoa_code','date_created2'))

head(aggregates1_3)
```

## Run a Moran Test

We can now run the model. This type of model is known as a Moran's test. This will create a correlation score between -1 and 1. Much like a correlation coefficient, 1 determines perfect positive spatial autocorrelation (so our data is clustered), 0 identifies the data is randomly distributed and -1 represents negative spatial autocorrelation (so dissimilar values are next  to each other).

```r
lsoa_aggrs2 = lsoa_aggrs[(which(!is.na(lsoa_aggrs$Total))),]
```

```r
# Creates local moran output
start_time = Sys.time()
local = localmoran(x = lsoa_aggrs$Total, listw = listw, zero.policy=TRUE, na.action = na.exclude)
print(paste("Time taken:", Sys.time()-start_time))
```

### Functions to generate Local LISA categories

```r
plot_layer = function(dataframe){
    library(tmap)
    map1 = tm_shape(dataframe) 
    map1 = map1 + tm_borders(alpha =0.4)
    map1 = map1 + tm_layout(main.title = "")
    map1    
}
```

```r
plot_LISA_map = function(dataframe_spatial, quadrant){
        brks = c(0,1,2,3,4)
        colors = c("white","blue",'lightblue1','orange',"red")
        plot(dataframe_spatial,border="lightgray",col=colors[findInterval(quadrant,brks,all.inside=FALSE)])
        box()
        legend("bottomleft"
               ,legend=c("insignificant"
                         ,"Variable less than mean-Behaviour different to neighbours"
                         ,"Variable less than mean-Behaviour similar to neighbours"
                         ,"Variable higher than mean-Behaviour different to neighbours"
                         ,"Variable higer than mean-Behaviour similar to neighbours"),
               fill=colors,bty="n")
}
```

```r
LISA_calculation = function(dataframe, variable, print_ = FALSE){
    
    if(print_ == TRUE){
        print('1. Calculating neighbours')
        print('----------------------------------------')    
    }
    
    start_time = Sys.time()
    neighbours = poly2nb(dataframe)
    neighbours_subset = subset(neighbours, subset = card(neighbours)>0)
    #print(paste("Time taken:", Sys.time()-start_time))
    
    if(print_ == TRUE){
        print('2. Convert neighbours to a list object')
        print('-----------------------------------------')
    }
    start_time = Sys.time()
    listw = nb2listw(neighbours)
    #print(paste("Time taken:", Sys.time()-start_time))
    
    if(print_ == TRUE){
        print('3. Run a local Moran test')
        print('-----------------------------------------')
    }        
    start_time = Sys.time()
    local = localmoran(x = variable, listw = listw, zero.policy=TRUE, na.action = na.exclude)
    #print(paste("Time taken:", Sys.time()-start_time))
    moran.map = cbind(dataframe, local)
    #print(tm_shape(moran.map) + tm_fill(col = "Ii", style = "quantile", title = "local moran statistic"))
    
    if(print_ == TRUE){
        print('4. Creation of LISA category')
        print('-----------------------------------------')
    }
    # Create a vector with all values = 0 of length equal to the number of polygons we have
    quadrant = vector(mode="numeric",length=nrow(local))
    
    # Centers the variable of interest around its mean
    m.total_crime = variable - mean(variable, na.rm=TRUE)  
    
    # Centers the local Moran's around the mean
    m.local = local[,1] - mean(local[,1], na.rm=TRUE)  
    
    # significance threshold
    signif = 0.2
    
    # builds a data quadrant
    # 4 = if area has higher proportion of variable than mean and their neighbours are similar
    # 3 = if area has higher proportion of variable than mean and their neighbours are different
    # 2 = if area has lower proportion of variable than mean and their neighbours are similar
    # 1 = if area has lower proportion of variable than mean and their neighbours are different
    quadrant[m.total_crime >0 & m.local>0] = 4  
    quadrant[m.total_crime <0 & m.local<0] = 1      
    quadrant[m.total_crime <0 & m.local>0] = 2
    quadrant[m.total_crime >0 & m.local<0] = 3
    quadrant[local[,5]>signif] = 0
    
    quadrant_cat = ifelse(quadrant==0,'Insignificant'
                      , ifelse(quadrant==1,'Lower than mean - Different to neighbours'
                              , ifelse(quadrant==2,'Lower than mean - Similar to neighbours'
                                    , ifelse(quadrant==3,'Higher than mean - Different to neighbours'
                                          , 'Higher than mean - Similar to neighbours'))))
    
    
    #plot_LISA_map(dataframe_spatial = dataframe, quadrant = quadrant)
    
    if(print_ == TRUE){
        print('5. Appending LISA category to dataframe')
        print('-----------------------------------------')
    }
    moran.map2 = moran.map
    moran.map2$local_lisa_category = quadrant_cat
    moran.map2$local_lisa_values = quadrant

    return(moran.map2)

    
}
```

# Hot Spot Analysis - using distances and Getis-Ord Gi

Another approach we can take is hot-spot analysis. The Getis-Ord Gi Statistic looks at neighbours within a defined proximity to identify where either high or low values cluster spatially. Here statistically significant hot-spots are recognised as areas of high values where other areas within a neighbourhood range also share high values too.

Whilst the spatial autocorrection considered units which shared borders, for Getis-Ord we are defining neighbours based on proximity. The example below shows the results where we have a search radius of 2000 metres.
